```python
category_mapping = {
    'food_services': ['fastfood', 'food', 'food_delivery'],
    'transportation': ['rideshare', 'auto','fuel'],
    'digital_services': ['entertainment', 'subscriptions', 'online_subscriptions', 'mobileapps','cable/phone'],
    'travel_and_accommodation': ['hotels', 'airline'],
    'wellness_services': ['health', 'personal care', 'gym'],
    'online_retail' : ['online_retail']
    }
# Apply the mapping to create a new 'grouped_category' column
df['grouped_category'] = df['merchantCategoryCode'].apply(lambda x: next((k for k, v in category_mapping.items() if x in v), x))
# Define a mapping of new categories based on frequency
top_categories = df['grouped_category'].value_counts().nlargest(5).index.tolist() 
category_mapping_freq = {cat: cat if cat in top_categories else 'other' for cat in df['grouped_category'].unique()}

# Apply the frequency-based mapping to create a new 'final_category' column
df['final_category'] = df['grouped_category'].map(category_mapping_freq)
# Drop the intermediate 'grouped_category' column if not needed
df.drop(columns=['grouped_category'], inplace=True)
```

```python
import numpy as np

def assign_city(country):
    if country == 'US':
        return np.random.choice(['New York', 'Los Angeles', 'Chicago', 'Houston', 'Miami'])
    elif country == 'MEX':
        return np.random.choice(['Mexico City', 'Monterrey', 'Guadalajara', 'Puebla', 'Cancun'])
    elif country == 'CAN':
        return np.random.choice(['Toronto', 'Vancouver', 'Montreal', 'Calgary', 'Ottawa'])
    elif country == 'PR':
        return np.random.choice(['San Juan', 'Ponce', 'Mayaguez', 'Bayamon', 'Caguas'])

df['City'] = df['acqCountry'].apply(assign_city)
```

$50000.0: 45-65 years (established, high earning adults)

$20000.0 & $15000.0: 35-55 years (mid-career professionals)

$10000.0 & $7500.0: 30-45 years (early to mid-career individuals)

$5000.0: 25-40 years (young professionals)

$2500.0 & $1000.0: 20-35 years (college students to young professionals)

$500 & $250.0: 18-25 years (young adults, possibly students)

```python
def assign_age(credit_limit):
    if credit_limit == 50000.0:
        return np.random.randint(45, 66)
    elif credit_limit in [20000.0, 15000.0]:
        return np.random.randint(35, 56)
    elif credit_limit in [10000.0, 7500.0]:
        return np.random.randint(30, 46)
    elif credit_limit == 5000.0:
        return np.random.randint(25, 41)
    elif credit_limit in [2500.0, 1000.0]:
        return np.random.randint(20, 36)
    elif credit_limit in [500, 250]:
        return np.random.randint(18, 26)

df['Age'] = df['creditLimit'].apply(assign_age)
bins = [17, 24, 34, 44, 54, 64, 75]  # Adjust bins as needed for your specific age group ranges
labels = ['18-24', '25-34', '35-44', '45-54', '55-64', '65+']  # Labels for the age groups
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=True)
```

```python
def assign_card_type(credit_limit):
    if credit_limit <= 1000:
        # Higher probability for student and secured cards for lower limits
        choices = ['student cards', 'secured cards', 'cash back cards', 'travel cards']
        probabilities = [0.5, 0.4, 0.1, 0.0]  # Adjust probabilities as per logical assumption
    elif credit_limit <= 5000:
        choices = ['student cards', 'secured cards', 'cash back cards', 'travel cards']
        probabilities = [0.2, 0.3, 0.4, 0.1]
    elif credit_limit <= 15000:
        choices = ['cash back cards', 'travel cards', 'student cards', 'secured cards']
        probabilities = [0.4, 0.4, 0.1, 0.1]
    else:
        # Higher limits likely to have cash back and travel cards
        choices = ['cash back cards', 'travel cards', 'student cards', 'secured cards']
        probabilities = [0.3, 0.6, 0.05, 0.05]

    return np.random.choice(choices, p=probabilities)

df['Card_Type'] = df['creditLimit'].apply(assign_card_type)
```
